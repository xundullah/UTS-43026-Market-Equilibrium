# Price Floor and Its Effect on Market Outcomes

This Jupyter Notebook analyzes the impact of a price floor on equilibrium.

## Problem Statement

Given:
- **Demand Function**: $ Q_d = 150 - 3P $
- **Supply Function**: $ Q_s = 40 + 2P $

A **government-imposed price floor** is set at $ P_f = 30 $, which is **above** the equilibrium price.

We want to:
1. Determine the **market equilibrium price and quantity** in the absence of the price floor.
2. Analyze the **surplus or shortage** created by the price floor.
3. Visualize the demand and supply curves, marking the price floor and the resulting **surplus or shortage**.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
def demand_function(price):
    return 150 - 3 * price

def supply_function(price):
    return 40 + 2 * price
```

```python
def find_equilibrium(price):
    return demand_function(price) - supply_function(price)

equilibrium_price = fsolve(find_equilibrium, 10)[0]
equilibrium_quantity = demand_function(equilibrium_price)

print(f'Equilibrium Price: {equilibrium_price:.2f}')
print(f'Equilibrium Quantity: {equilibrium_quantity:.2f}')
```

```python
price_floor = 30
quantity_demanded_at_floor = demand_function(price_floor)
quantity_supplied_at_floor = supply_function(price_floor)
surplus = quantity_supplied_at_floor - quantity_demanded_at_floor

print(f'Surplus Due to Price Floor: {surplus}')
```

```python
prices = np.linspace(0, 50, 100)
plt.figure(figsize=(10, 6))
plt.plot(prices, demand_function(prices), label='Demand Curve', color='blue')
plt.plot(prices, supply_function(prices), label='Supply Curve', color='red')
plt.axvline(x=price_floor, linestyle='--', color='gray', label='Price Floor')
plt.scatter(price_floor, quantity_demanded_at_floor, color='blue', label='Quantity Demanded at Floor')
plt.scatter(price_floor, quantity_supplied_at_floor, color='red', label='Quantity Supplied at Floor')
plt.title('Effect of Price Floor on Market')
plt.xlabel('Price (P)')
plt.ylabel('Quantity (Q)')
plt.legend()
plt.grid(True)
plt.show()
```
