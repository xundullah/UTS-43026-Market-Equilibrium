# Tax Impact on Market Equilibrium

This Jupyter Notebook demonstrates how a per-unit tax affects market equilibrium, causing changes in price, quantity, and deadweight loss.

## Problem Statement

Given:
- **Demand Function**: $ Q_d = 120 - 4P $
- **Supply Function**: $ Q_s = 30 + 2P $

A per-unit tax of $t = 5$ is introduced, shifting the supply function to:
$ Q_s = 30 + 2(P - t) $

We want to:
1. Find the new equilibrium price and quantity after the tax.
2. Compute the deadweight loss caused by the tax.
3. Visualize the demand and supply curves before and after the tax.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
def demand_function(price):
    return 120 - 4 * price

def supply_function(price):
    return 30 + 2 * price

def supply_function_with_tax(price, tax=5):
    return 30 + 2 * (price - tax)
```

```python
def find_equilibrium(price):
    return demand_function(price) - supply_function(price)

def find_equilibrium_with_tax(price):
    return demand_function(price) - supply_function_with_tax(price)

equilibrium_price = fsolve(find_equilibrium, 10)[0]
equilibrium_quantity = demand_function(equilibrium_price)

equilibrium_price_tax = fsolve(find_equilibrium_with_tax, 10)[0]
equilibrium_quantity_tax = demand_function(equilibrium_price_tax)

print(f'Equilibrium Price before Tax: {equilibrium_price:.2f}')
print(f'Equilibrium Quantity before Tax: {equilibrium_quantity:.2f}')
print(f'Equilibrium Price after Tax: {equilibrium_price_tax:.2f}')
print(f'Equilibrium Quantity after Tax: {equilibrium_quantity_tax:.2f}')
```

```python
deadweight_loss = 0.5 * 5 * (equilibrium_quantity - equilibrium_quantity_tax)
print(f'Deadweight Loss: {deadweight_loss:.2f}')
```

```python
prices = np.linspace(0, 50, 100)
plt.figure(figsize=(10, 6))
plt.plot(prices, demand_function(prices), label='Demand Curve', color='blue')
plt.plot(prices, supply_function(prices), label='Supply Curve', color='red')
plt.plot(prices, supply_function_with_tax(prices), label='Supply Curve with Tax', linestyle='dashed', color='orange')
plt.scatter(equilibrium_price, equilibrium_quantity, color='black', label='Equilibrium (Before Tax)')
plt.scatter(equilibrium_price_tax, equilibrium_quantity_tax, color='green', label='Equilibrium (After Tax)')
plt.axvline(x=equilibrium_price, linestyle='--', color='gray')
plt.axvline(x=equilibrium_price_tax, linestyle='--', color='black')
plt.title('Impact of Tax on Market Equilibrium')
plt.xlabel('Price (P)')
plt.ylabel('Quantity (Q)')
plt.legend()
plt.grid(True)
plt.show()
```
