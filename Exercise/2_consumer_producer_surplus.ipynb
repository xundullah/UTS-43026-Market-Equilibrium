# Consumer and Producer Surplus Calculation

This Jupyter Notebook calculates consumer and producer surplus at equilibrium.

## Problem Statement

Given:
- **Demand Function**: $ Q_d = 200 - 5P $
- **Supply Function**: $ Q_s = 50 + 4P $

We want to:
1. Determine the equilibrium price and quantity.
2. Compute the **consumer surplus** and **producer surplus**.
3. Plot the demand and supply curves, shading surplus areas.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
def demand_function(price):
    return 200 - 5 * price

def supply_function(price):
    return 50 + 4 * price
```

```python
def find_equilibrium(price):
    return demand_function(price) - supply_function(price)

equilibrium_price = fsolve(find_equilibrium, 10)[0]
equilibrium_quantity = demand_function(equilibrium_price)

print(f'Equilibrium Price: {equilibrium_price:.2f}')
print(f'Equilibrium Quantity: {equilibrium_quantity:.2f}')
```

```python
max_price_willing_to_pay = 40  # When Qd = 0
min_price_willing_to_accept = -12.5  # When Qs = 0

consumer_surplus = 0.5 * (max_price_willing_to_pay - equilibrium_price) * equilibrium_quantity
producer_surplus = 0.5 * (equilibrium_price - min_price_willing_to_accept) * equilibrium_quantity

print(f'Consumer Surplus: {consumer_surplus:.2f}')
print(f'Producer Surplus: {producer_surplus:.2f}')
```

```python
prices = np.linspace(0, 50, 100)
plt.figure(figsize=(10, 6))
plt.plot(prices, demand_function(prices), label='Demand Curve', color='blue')
plt.plot(prices, supply_function(prices), label='Supply Curve', color='red')
plt.axhline(y=equilibrium_quantity, linestyle='--', color='gray')
plt.axvline(x=equilibrium_price, linestyle='--', color='black')
plt.fill_between(prices, demand_function(prices), equilibrium_quantity, where=(prices < equilibrium_price), color='blue', alpha=0.2, label='Consumer Surplus')
plt.fill_between(prices, supply_function(prices), equilibrium_quantity, where=(prices < equilibrium_price), color='red', alpha=0.2, label='Producer Surplus')
plt.title('Consumer and Producer Surplus')
plt.xlabel('Price (P)')
plt.ylabel('Quantity (Q)')
plt.legend()
plt.grid(True)
plt.show()
```
